# Treating Missing Values

Filling missing values using fillna(), replace() and interpolate()

```python
#data_creation with missing values
data={'names':['kamal','ram','suraj','mahesh','bharat','abhishek','rocky'],
      'age':[26,27,28,30,29,21,21],
      'gender':['M','M','M','M','M','F','M'],
      'rank':[6,5,4,3,2,1,0]}
rank_data=DataFrame(data)
rank_data.iloc[2:4,1]=np.nan
rank_data.iloc[[5],[1]]=np.nan
rank_data.iloc[1:3,3]=np.nan
rank_data.iloc[6,:]=np.nan

rank_data
```

```python
rank_data.isnull()
```

```python
rank_data.notnull()
```

```python
bool_series=pd.isnull(rank_data['age'])
rank_data[bool_series]
```

```python
#filling missing value
rank_data.fillna(0)
```

```python
rank_data.fillna(method='pad') #missing value are filled with value comes prior to them
```

```python
rank_data.fillna(method='bfill') #missing value are filled with value comes after them
```

```python
rank_data
```

```python
rank_data.interpolate(method='linear') #missing value are filled with interpolated values prior and after them
```

```python
rank_data.dropna() #only that row is dropped that has a missing value
```

```python
rank_data.dropna(how='all') #only that row is dropped that has all missing value
```

```python
rank_data.dropna(axis=1) #only that column is dropped that has a missing value; all columns have atleast one missing value
```

```python
rank_data.dropna(axis=0) #only that row is dropped that has a missing value
```

## Concatenating and transforming

Concatenating Data

```python
df_new= DataFrame(np.arange(36).reshape(6,6))
df_new
```

```python
df_new_2=DataFrame(np.arange(28).reshape(4,7))
df_new_2
```

```python
pd.concat([df_new, df_new_2], axis=1) #concatenating along columns (side-by-side)
```

```python
pd.concat([df_new, df_new_2]) # concatenating along rows (stack on top of other)
```

Transforming data

Dropping data

```python
df_new.drop([0,2]) #drop rows
```

```python
df_new_drop=df_new.drop([0,2], axis=1) #drop columns
df_new_drop #can be used in dropping certain features
#df_new_drop=df_new.drop(['column-name','column-name2',axis=1]) can also be used to drop columns with certain names
```

```python
df_new_drop.columns=['Column1','Column2','Column3','Column4']
df_new_drop #renaming the column index
```

Addding data

```python
series_obj=Series(np.arange(6))
series_obj.name='added_variable'
series_obj #adding new feature
```

```python
df_added=DataFrame.join(df_new_drop, series_obj)
df_added
```

```python
df_added_con=pd.concat([df_added, df_added], ignore_index=True)
df_added_con
```

Sorting data

```python
df_sorted=df_added.sort_values(by=['Column3'],ascending=[False])
df_sorted
```
